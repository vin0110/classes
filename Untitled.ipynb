```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(1,40,100)
```

```python
stages = [2, 4, 8, 16, 32]
for n in stages:
    y = x / (n + x - 1)
    plt.plot(x, y, label=str(n))

#plt.plot(x, x/x, '--')

plt.legend()
plt.ylabel('Instructions/second (x10^9)')
plt.xlabel("Runlength")
plt.title("Pipeline length effect on performance")
plt.savefig('perf1.png')
```

```python
cycles_per_second = 1
stages = [2, 4, 8, 16, 32]
for n in stages:
    y = cycles_per_second * x * n / (n + x - 1)
    plt.plot(x, y, label=str(n))

plt.plot(x, x/x, '--')

plt.legend()
plt.ylabel('Instructions/second (x10^9)')
plt.xlabel("Runlength")
plt.title("Pipeline length effect on performance")
plt.savefig('perf2.png')
plt.show()
```

```python
cycles_per_second = 1
stages = [2, 4, 8, 16, 32]
for n in stages:
    y = cycles_per_second * x * (np.log(n)*.8+1) / (n + x - 1)
    plt.plot(x, y, label=str(n))

plt.plot(x, x/x, '--')

plt.legend(loc='lower right')
plt.ylabel('Instructions/second (x10^9)')
plt.xlabel("Runlength")
plt.title("Pipeline length effect on performance")
plt.savefig('perf3.png')
plt.show()
```

```python
k = [1] + stages
clock = [(np.log(s)*.8 + 1) for s in k]
#.ylim(0.)
plt.plot(k, clock, 'o')
plt.title('Clock vs. stages')
plt.ylabel('GHz')
plt.xlabel('Stages')
plt.savefig('clock.png')
```
