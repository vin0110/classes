```python
import pandas as pd
import numpy as np
import json
import glob
%matplotlib inline  

import matplotlib.pyplot as plt 
from datetime import datetime
file_path = 'C:/Users/Prerona/Desktop/CSept/'
#file_path = 'C:/Users/Prerona/Desktop/CSept/C68U.csv'
for file_name in glob.glob(file_path+'*.csv'):
    df = pd.read_csv(file_name, header=None, usecols=[0, 1, 4, 5, 6])

    def convertToDate(date_str):
        year = str(date_str)[:2]
        if int(year) > 50:
            year = '19'+year
        else:
            year = '20'+year

        return pd.to_datetime(str(date_str)[2:4]+"/"+str(date_str)[4:]+"/"+year)

    def days_between(d1):
        enddate = df["date"].iloc[-1]
        return abs(enddate - d1).days

    df['date'] = df[0].apply(convertToDate)

    df['days_diff'] = df["date"].apply(days_between)
    df = df.drop(0, axis=1)
    df = df.rename(columns={1: "open", 4: "close", 5: "volume", 6: "openint"})
    df_copy = df.copy()
    #print(df_copy)
    #print(df["days_diff"].iloc[0])
    days_max= df["days_diff"].iloc[0]

    max = days_max
    time_series = []
    close_series = []
    volume_series = []
    openint_series = []

    for i in range(len(df_copy.index)):
        arr = df_copy.iloc[i].as_matrix()
        if arr[5] == max:
            time_series.append(arr[5])
            close_series.append(arr[1])
            volume_series.append(arr[2])
            openint_series.append(arr[3])
            max = max-1
        elif arr[5] < max:
            #prev_time_value =  time_series[-1]
            prev_closeprice_value =  close_series[-1]
            prev_volume_value =  volume_series[-1]
            prev_openint_value =  openint_series[-1]
            while max > arr[5]:
                time_series.append(max)
                close_series.append(prev_closeprice_value)
                volume_series.append(prev_volume_value)
                openint_series.append(prev_openint_value)
                max = max - 1
            time_series.append(arr[5])
            close_series.append(arr[1])
            volume_series.append(arr[2])
            openint_series.append(arr[3])
            max = max-1

    #generating plot of the Closing prices against the time
    plt.plot(time_series, close_series)
    #plt.plot(time_series, volume_series, label = "Volume")
    #plt.plot(time_series, openint_series, label = "Open Interest")
    plt.xlabel("Days from the due date")
    plt.ylabel("Closing prices")
            
    
```

```python
import pandas as pd
import numpy as np
import json
import glob
%matplotlib inline  

import matplotlib.pyplot as plt 
from datetime import datetime
from scipy.stats import pearsonr
pricebefore50 = []
pricebefore100 = []
pricebefore150 = []
duedateprices = []
file_path = 'C:/Users/Prerona/Desktop/CSept/'
#file_path = 'C:/Users/Prerona/Desktop/CSept/C68U.csv'
for file_name in glob.glob(file_path+'*.csv'):
    df = pd.read_csv(file_name, header=None, usecols=[0, 1, 4, 5, 6])

    def convertToDate(date_str):
        year = str(date_str)[:2]
        if int(year) > 50:
            year = '19'+year
        else:
            year = '20'+year

        return pd.to_datetime(str(date_str)[2:4]+"/"+str(date_str)[4:]+"/"+year)

    def days_between(d1):
        enddate = df["date"].iloc[-1]
        return abs(enddate - d1).days

    df['date'] = df[0].apply(convertToDate)

    df['days_diff'] = df["date"].apply(days_between)
    df = df.drop(0, axis=1)
    df = df.rename(columns={1: "open", 4: "close", 5: "volume", 6: "openint"})
    df_copy = df.copy()
    #print(df_copy)
    #print(df["days_diff"].iloc[0])
    days_max= df["days_diff"].iloc[0]

    max = days_max
    time_series = []
    close_series = []
    volume_series = []
    openint_series = []

    for i in range(len(df_copy.index)):
        arr = df_copy.iloc[i].as_matrix()
        if arr[5] == max:
            time_series.append(arr[5])
            close_series.append(arr[1])
            volume_series.append(arr[2])
            openint_series.append(arr[3])
            max = max-1
        elif arr[5] < max:
            #prev_time_value =  time_series[-1]
            prev_closeprice_value =  close_series[-1]
            prev_volume_value =  volume_series[-1]
            prev_openint_value =  openint_series[-1]
            while max > arr[5]:
                time_series.append(max)
                close_series.append(prev_closeprice_value)
                volume_series.append(prev_volume_value)
                openint_series.append(prev_openint_value)
                max = max - 1
            time_series.append(arr[5])
            close_series.append(arr[1])
            volume_series.append(arr[2])
            openint_series.append(arr[3])
            max = max-1
    
    duedateprices.append(close_series[-1])
    
    pricebefore50.append(close_series[-51])
    
    if len(close_series)>100:
        pricebefore100.append(close_series[-101])
    else:
        pricebefore100.append(close_series[0])
        
    if len(close_series)>150:
        pricebefore150.append(close_series[-151])
    else:
        pricebefore150.append(close_series[0])

corr50, _ = pearsonr(duedateprices, pricebefore50)
corr100, _ = pearsonr(duedateprices, pricebefore100)
corr150, _ = pearsonr(duedateprices, pricebefore150)
print('Pearsons correlation: %.3f' % corr50)
print('Pearsons correlation: %.3f' % corr100)
print('Pearsons correlation: %.3f' % corr150)
```
